# Jumlah Kalori Berdasarkan Item

```python
brands = df.groupby('Item').count()[['Protein']].sort_values(by='Protein', ascending=True).reset_index()
brands = brands.rename(columns={'Protein':'Calories'})
```

```python
fig = plt.figure(figsize = (10,7))
sns.barplot(x = brands['Item'], y = brands['Calories'], color = 'red')
plt.xticks(rotation = 60)
```

# Seleksi Fitur

```python
fitur = ['Cholesterol', 'Total_Carb', 'Sugars', 'Protein']
x = df[fitur]
y = df['Calories']
x.shape, y.shape
```

# Split Data Training & Testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 70)
y_test.shape
```

# membuat regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
predik = lr.predict(X_test)
```

```python
y = lr.score(X_test, y_test)
print('Akurasi Model Regresi Linier : ', y)
```

# Membuat inputan regresi linier

```python
#Cholesterol (mg) = 255 , Total_Carb (g) = 49 , Sugars (g) = 11 , Protein (g) = 67
inputan = np.array([[255, 49, 11, 67]])
prediksi = lr.predict(inputan)
print('Estimasi Jumlah Kalori Menu Burger King : ', prediksi)
```

# Save Model

```python
import pickle
filename = 'estimasi_burgerking.sav'
pickle.dump(lr,open(filename,'wb'))
```
